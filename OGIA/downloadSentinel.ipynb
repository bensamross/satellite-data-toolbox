```python
import openeo
from shapely.geometry import box
```

```python
connection = openeo.connect(url="openeo.dataspace.copernicus.eu")
connection.authenticate_oidc()
```

```python
# Define area of interest (bounding box: minx, miny, maxx, maxy)
spatial_extent = {
    "west": 152.8,
    "south": -27.7,
    "east": 153.0,
    "north": -27.5,
    "crs": "EPSG:4326"
}

# Define time range
start_date = "2023-06-01"
end_date = "2023-06-30"

# Load Sentinel-2 L2A data
datacube = connection.load_collection(
    "SENTINEL2_L2A",
    spatial_extent=spatial_extent,
    temporal_extent=[start_date, end_date],
    bands=["B04", "B08"]  # Red and NIR bands
)

# Example: Calculate NDVI
ndvi = datacube.ndvi(nir="B08", red="B04")

# Save as GeoTIFF locally using a batch job
job = connection.create_job(ndvi.save_result(format="GTiff"))
job.start_and_wait()
job.download_results("sentinel2_download")
```
