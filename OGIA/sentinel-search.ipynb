```python
import json
import requests
import pandas as pd
```

```python
json = requests.get("https://catalogue.dataspace.copernicus.eu/resto/api/collections/search.json?").json()
pd.DataFrame.from_dict(json['features']).head(3)
```

```python
json = requests.get("https://catalogue.dataspace.copernicus.eu/resto/api/collections/Sentinel2/search.json?productType=S2MSI1C&cloudCover=[0,10]&startDate=2022-06-11T00:00:00Z&completionDate=2022-06-22T23:59:59Z&maxRecords=10&box=152.75,-27.75,153.25,-27.25").json()
pd.DataFrame.from_dict(json['features']).head(3)
```

```python
import folium

# Extract the geometry from the first feature
geometry = json['features'][0]['geometry']

# Create a folium map centered at the first coordinate of the geometry
coordinates = geometry['coordinates'][0][0]  # Assuming it's a Polygon
m = folium.Map(location=[coordinates[1], coordinates[0]], zoom_start=8)

# Add the geometry to the map
folium.GeoJson(geometry).add_to(m)

# Display the map
m
```
